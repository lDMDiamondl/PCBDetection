```python
import kaggle
import os

dataset_name = 'akhatova/pcb-defects'
download_path = './data'
```

```python
# 데이터가 없을 때만 다운로드 (중복 방지)

if not os.path.exists(download_path) or len(os.listdir(download_path)) == 0:
    os.makedirs(download_path, exist_ok=True)
    try:
        kaggle.api.dataset_download_files(dataset_name, path=download_path, unzip=True)
        print("다운로드 완료")
    except Exception as e:
        print(f"다운로드 실패: {e}")
else:
    print("다운로드 스킵")

# 폴더 구조 출력 + root 찾기
target_root = None

for root, dirs, files in os.walk(download_path):
    level = root.replace(download_path, '').count(os.sep)
    indent = ' ' * 4 * level
    print(f"{indent}{os.path.basename(root)}/")

    # 핵심 조건 (PCB 전용)
    if 'Annotations' in dirs and 'images' in dirs:
        target_root = root

print("\n detected dataset root:", target_root)
```
